```python
# 5- Create a new column named average_per_backer and set its value to the total amount pledged / number of backers.
df_ks['average_per_backer'] = df_ks['usd_pledged'] / df_ks['backers']
df_ks.head(20)
```

```python
# 6- What happened to the rows with 0 backers? How can this be dealt with?
df_ks['average_per_backer'] = df_ks.apply(lambda row: np.inf if row['backers'] == 0 else row['usd_pledged'] / row['backers'], axis = 1)
df_ks.head(20)
```

```python
# 6- What happened to the rows with 0 backers? How can this be dealt with?
def average_calc(row):
  try:
    return row['usd_pledged'] / row['backers']
  except ZeroDivisionError:
    return np.inf
df_ks['average_per_backer'] = df_ks.apply(average_calc, axis =1)
df_ks.head(20)
```

```python
# 6- What happened to the rows with 0 backers? How can this be dealt with?
df_ks.loc[df_ks['backers'] == 0, 'average_per_backer'] = np.inf
df_ks.loc[df_ks['backers'] != 0, 'average_per_backer'] = df_ks['usd_pledged'] / df_ks['backers']
df_ks.head(20)
```

```python
# 6- What happened to the rows with 0 backers? How can this be dealt with?
df_ks['average_per_backer'] = np.where(df_ks['backers'] == 0, np.nan, df_ks['usd_pledged'] / df_ks['backers'])
```

```python
# 8- Create a DataFrame whose index is the first 10 letters of the alphabet
# and contains two more columns with first 10 prime numbers and the first 10 fibonacci numbers.
indices = 'a b c d e f g h i j'.split()
prime_numbers = []
for i in range(2,30):
  for j in range(2,i):
    if i % j == 0:
      break
  else:
    prime_numbers.append(i)

fibonacci_numbers = []
for i in range(0,10):
  if i < 2:
    fibonacci_numbers.append(i)
  else:
    fibonacci_numbers.append(fibonacci_numbers[i-1] + fibonacci_numbers[i-2])

dict_prime_fib = {'prime_numbers' : prime_numbers, 'fibonacci_numbers' : fibonacci_numbers}
df_prime_fib = pd.DataFrame(dict_prime_fib, index = indices)
df_prime_fib

arr_prime_fib = np.array([prime_numbers, fibonacci_numbers])
df_prime_fib_2 = pd.DataFrame(arr_prime_fib.T, index = indices, columns = ['Prime', 'Fibo'])
df_prime_fib_2
```

```python
# 11- Merge these two DataFrames into a single DataFrame called census.
# Please use the following dataset for merging purposes:
df_census_div = pd.read_csv('/content/census-divisions.csv')
df_pop = pd.read_csv('/content/state-populations.csv')
census = pd.merge(df_pop, df_census_div, on = 'state', how = 'left')
```

```python
# 12- Re-shape census so that one column contains all population measures, and another the year attributes.
census.head()
census_melt = pd.melt(census, id_vars =['state', 'region', 'division'], var_name = 'year', value_name = 'Population')
```
